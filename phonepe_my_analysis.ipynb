# Task 3: Data Quality Checks
3.1: Ensure data consistency across state and district levels

1. For each state, calculate the total number of transactions, total transaction amount, and total registered users by summing up the values from the district level data.
2. Compare the results with the corresponding values at the state level to ensure they match.
3. Display any discrepancies found between the district-level and state-level data.

# To perform data quality checks and ensure data consistency across state and district levels, we will :-

1. Computing the complete number of transactions, total transaction amount, and total registered users for each state by summarizing the values from the district-level data.

2. Comparing the outcomes with corresponding values at the 'state level'.

3. Showing any disparities found between the the district-level and state-level data.

```python
# Well here is the code to execute these steps:

# Calculating the total number of transactions, total transaction amount, and total registered users for each state from district-level data
district_aggregated_data = district_txn_users.groupby('State').agg(
    District_Total_Transactions=('Transactions', 'sum'),
    District_Total_Amount=('Amount (INR)', 'sum'),
    District_Total_Registered_Users=('Registered Users', 'sum')
).reset_index()

# Aggregating state-level data to match the structure of district-aggregated data
state_aggregated_data = state_txn_users.groupby('State').agg(
    State_Total_Transactions=('Transactions', 'sum'),
    State_Total_Amount=('Amount (INR)', 'sum'),
    State_Total_Registered_Users=('Registered Users', 'sum')
).reset_index()

# Merging the aggregated data to compare district-level and state-level data
comparison_data = pd.merge(state_aggregated_data, district_aggregated_data, on='State', how='outer', suffixes=('_State', '_District'))

# Checking for discrepancies
discrepancies = comparison_data[
    (comparison_data['State_Total_Transactions'] != comparison_data['District_Total_Transactions']) |
    (comparison_data['State_Total_Amount'] != comparison_data['District_Total_Amount']) |
    (comparison_data['State_Total_Registered_Users'] != comparison_data['District_Total_Registered_Users'])
]

print("Discrepancies found between district-level and state-level data:")
print(discrepancies)
```

# 4.1.1: Merge the State_Txn and Users dataset with the District Demographics dataset

First, we need to aggregate the population data at the state level from the District Demographics dataset.

Then, we merge this aggregated data with the State_Txn and Users dataset.

```python
# Aggregating population data at the state level :-
state_population = district_demographics.groupby('State').agg(
    Total_Population=('Population', 'sum')
).reset_index()

# Merging the State_Txn and Users dataset with the aggregated state population data
state_user_population_ratio = pd.merge(state_txn_users, state_population, on='State', how='left')

# Calculating the ratio of registered users to the population for each state
state_user_population_ratio['User_Population_Ratio'] = state_user_population_ratio['Registered Users'] / state_user_population_ratio['Total_Population']

# Displaying the results
state_user_population_ratio_table = state_user_population_ratio[['State', 'User_Population_Ratio']]
print(state_user_population_ratio_table)
```

```python
import matplotlib.pyplot as plt

# Plotting the ratio of users to population by 'state' :-

plt.figure(figsize=(12, 8))
plt.barh(state_user_population_ratio_table['State'], state_user_population_ratio_table['User_Population_Ratio'])
plt.xlabel('User to Population Ratio')
plt.ylabel('State')
plt.title('Ratio of Users to Population by State')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading data :-
file_path = 'phonepe-pulse_raw-data_q12018-to-q22021-v0-1-5-1720351752.xlsx'
district_demographics_df = pd.read_excel(file_path, sheet_name='District Demographics')

# Selecting a state
selected_state = 'Karnataka'  # Replace with your selected state
state_district_data = district_demographics_df[district_demographics_df['State'] == selected_state]

# Creating bar plot -
plt.figure(figsize=(12, 6))
plt.bar(state_district_data['District'], state_district_data['Density'])
plt.xlabel('District')
plt.ylabel('Population Density')
plt.title(f'Population Density of Districts in {selected_state}')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# Task 6.1: Identify any trends or patterns in the transaction data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data
file_path = 'phonepe-pulse_raw-data_q12018-to-q22021-v0-1-5-1720351752.xlsx'
state_txn_users_df = pd.read_excel(file_path, sheet_name='State_Txn and Users')

# Grouping data by year and quarter to identify trends :-
grouped_data = state_txn_users_df.groupby(['Year', 'Quarter']).sum().reset_index()

# Plotting total transactions over time :-
plt.figure(figsize=(12, 6))
plt.plot(grouped_data['Year'].astype(str) + 'Q' + grouped_data['Quarter'].astype(str), grouped_data['Transactions'], marker='o')
plt.xlabel('Time (Year and Quarter)')
plt.ylabel('Total Transactions')
plt.title('Total Transactions Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plotting total amount over time :-
plt.figure(figsize=(12, 6))
plt.plot(grouped_data['Year'].astype(str) + 'Q' + grouped_data['Quarter'].astype(str), grouped_data['Amount (INR)'], marker='o')
plt.xlabel('Time (Year and Quarter)')
plt.ylabel('Total Amount (INR)')
plt.title('Total Amount Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()





# FINDINGS :-

# 1. Influence of External Events:      observe any unusual spikes or dips in the trend lines.
#                                       These could correlate with external events like economic changes, policy announcements, or even the COVID-19 pandemic.

# 2. Overall Growth:                  We can see a clear upward trend in both "total transactions" and "transaction amount" over the years,
#                                      which indicates significant growth in digital payments adoption.

# 3. State-wise Differences:          While this analysis looks at overall trends, remember that individual states might show different patterns.


# 4. Seasonal Variations:             We might see some irregularity in the data, with potential spikes in transactions during festive seasons or specific quarters.
#                                     Further analysis can pinpoint these periods.
```

# Task 6.2: Correlate demographic data with transaction data

```python
import pandas as pd

# Loading the data :-
file_path = 'phonepe-pulse_raw-data_q12018-to-q22021-v0-1-5-1720351752.xlsx'
district_txn_users_df = pd.read_excel(file_path, sheet_name='District_Txn and Users')
district_demographics_df = pd.read_excel(file_path, sheet_name='District Demographics')

# Merging the dataframes on State and District :-
merged_data = pd.merge(district_txn_users_df, district_demographics_df, on=['State', 'District'])

# Calculating the correlation matrix :-
correlation_matrix = merged_data[['Population', 'Density', 'Transactions', 'Amount (INR)']].corr()

print("Correlation Matrix:")
print(correlation_matrix)
```
